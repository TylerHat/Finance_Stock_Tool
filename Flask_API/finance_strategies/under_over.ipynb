```python
#load the data
import yfinance as yf
import pandas as pd

# Function to get stock information
def get_stock_info(ticker):
    stock = yf.Ticker(ticker)
    try:
        pe_ratio = stock.info.get('forwardPE', None)
    except KeyError:
        pe_ratio = None
    try:
        current_price = stock.info.get('currentPrice', None)
    except KeyError:
        current_price = None
    try:
        eps = stock.info.get('trailingEps', None)
    except KeyError:
        eps = None
    return ticker, pe_ratio, current_price, eps

# Placeholder for the DataFrame
df_list = []

# Get the list of tickers on the NYSE (adjust the source as needed)
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].tolist()

# Get information for each ticker and append to DataFrame list
for ticker in tickers:
    ticker, pe_ratio, current_price, eps = get_stock_info(ticker)
    df_list.append({'Ticker': ticker, 'PE_ratio': pe_ratio, 'Current_Price': current_price, 'Earnings_per_share': eps})

# Convert the list to DataFrame
df = pd.DataFrame(df_list)

df
```

```python
#calculater and show the mean P/E ratio

pe_ratio_mean = df.PE_ratio.mean()

pe_ratio_mean
```

```python
#calculate the fair market value
#if Fair Market Value is higher than current price then stock is undervalued
df['Fair_Market_Value'] = pe_ratio_mean * df['Earnings_per_share']

df
```

```python
#Calculate and show the value ratio to determined if company ifs over valued or undervalued
df['Over_Under_Ratio'] = df['Current_Price'] / df['Fair_Market_Value']

df
```

```python
#create a new column to display a label of under or over valued 
df['Value_Label'] = np.where(df['Over_Under_Ratio'] <1.0, 'Under Valued', 'Fair or Over Valued')

df
```

```python
#Percent overvalued calc
df['Value_Percent'] = abs(df['Over_Under_Ratio']-1)*100

df
```

```python
#show only a list of stocks that are concidered under valued
df[df.Value_Label =='Under Valued'].Ticker
```
